## EDA:  temp_usage_charging_effect_realistic_data dataset

```python
import glob
import pandas as pd
import os

# Define the path to the "synthetic data" folder
synthetic_data_folder = '../synthetic_data/temp_usage_charging_effect_realistic_data'  # Adjust the path as necessary

# Use glob to find all CSV files in the specified folder
csv_files = glob.glob(os.path.join(synthetic_data_folder, '*_battery_data.csv'))

# Load each CSV file into a dictionary of DataFrames
dataframes = {os.path.basename(f).split('_battery_data.csv')[0]: pd.read_csv(f) for f in csv_files}
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Convert 'Date' column to datetime
for df in dataframes.values():
    df['Date'] = pd.to_datetime(df['Date'])

# Set up the plotting style
#plt.style.use('seaborn')
#sns.set_palette("deep")

# Function to plot SoH over time for all scenarios
def plot_soh_over_time():
    plt.figure(figsize=(15, 10))
    for name, df in dataframes.items():
        plt.plot(df['Date'], df['SOH'], label=name)
    plt.title('State of Health (SoH) Over Time')
    plt.xlabel('Date')
    plt.ylabel('State of Health (%)')
    plt.legend()
    plt.grid(True)
    plt.show()

# Function to plot temperature distribution
def plot_temperature_distribution():
    plt.figure(figsize=(15, 10))
    for name, df in dataframes.items():
        sns.kdeplot(df['Temperature'], label=name, shade=True)
    plt.title('Temperature Distribution')
    plt.xlabel('Temperature (°C)')
    plt.ylabel('Density')
    plt.legend()
    plt.grid(True)
    plt.show()

# Function to plot correlation heatmap
def plot_correlation_heatmap(df_name):
    plt.figure(figsize=(12, 10))
    sns.heatmap(dataframes[df_name].corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1, center=0)
    plt.title(f'Correlation Heatmap - {df_name}')
    plt.show()

# Function to plot SoH vs Total Cycles
def plot_soh_vs_cycles():
    plt.figure(figsize=(15, 10))
    for name, df in dataframes.items():
        plt.scatter(df['Total_Cycles'], df['SOH'], alpha=0.5, label=name)
    plt.title('State of Health (SoH) vs Total Cycles')
    plt.xlabel('Total Cycles')
    plt.ylabel('State of Health (%)')
    plt.legend()
    plt.grid(True)
    plt.show()

# Function to plot boxplot of charging behaviors
def plot_charging_behaviors():
    plt.figure(figsize=(15, 10))
    data = []
    labels = []
    for name, df in dataframes.items():
        data.append(df['Charge_End'] - df['Charge_Start'])
        labels.append(name)
    plt.boxplot(data, labels=labels)
    plt.title('Charging Behavior Comparison')
    plt.xlabel('Scenario')
    plt.ylabel('Charge Range (%)')
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.show()

# Execute the plotting functions
plot_soh_over_time()
plot_temperature_distribution()
plot_soh_vs_cycles()
plot_charging_behaviors()

# Plot correlation heatmap for each scenario
for name in dataframes.keys():
    plot_correlation_heatmap(name)

# Calculate and print summary statistics
for name, df in dataframes.items():
    print(f"\nSummary Statistics for {name}:")
    print(df.describe())
    print("\nFinal SoH:", df['SOH'].iloc[-1])
    print("Total Cycles:", df['Total_Cycles'].iloc[-1])
    print("Average Daily Mileage:", df['Daily_Mileage'].mean())
    print("Average Temperature:", df['Temperature'].mean())
```
